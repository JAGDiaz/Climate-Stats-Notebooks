```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import cluster
```

```python
styledict = {'xtick.labelsize':25,
             'xtick.major.size':9,
             'xtick.major.width':1,
             'ytick.labelsize':25,
             'ytick.major.size':9,
             'ytick.major.width':1,
             'legend.framealpha':0.0,
             'legend.fontsize':15,
             'axes.labelsize':20,
             'axes.titlesize':25,
             'axes.linewidth':2,
             'figure.figsize':(12,8),
             'savefig.format':'jpg'}

plt.rcParams.update(**styledict)
```

### K-means clustering

```python
mydata = np.random.uniform(size=(250,2))

kmeans = cluster.KMeans(n_clusters=5, random_state=0)
y_mean = kmeans.fit_predict(mydata)
centers = kmeans.cluster_centers_

colors = ['black', 'red', 'green', 'blue', 'purple']
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12,12))

for (i, cent, colour) in zip(range(len(colors)), centers, colors):
    ax.plot(cent[0], cent[1], color=colour, marker='*', ms=12)
    ax.plot(mydata[y_mean==i,0], mydata[y_mean==i,1], color=colour,
            marker='o', ms=8, ls='', label=f"Cluster {i+1}")

ax.set_ylim(-0.05, 1.1)
ax.set_xlabel("$x_1$")
ax.set_ylabel("$x_2$")
ax.legend(loc='upper center', ncol=5)

ax.set_title("K-means clustering of 250 uniformly random points")

fig.savefig("CH12;K-MeansClustering.jpg", bbox_inches='tight')
```
