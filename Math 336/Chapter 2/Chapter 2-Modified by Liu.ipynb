## Order of Operations

The order in which operations are executed in a single expression follow the standard rules of mathematics, for example:

```python
print(2*3**5 + 1) # First the exponentiation is done, then the multiplcation, and lastly the addition.
print((2*(3**5)) + 1) # This set of operations is equivalent to the set above, but with added parentheses to
                      # emphasize the order.
```

If, instead, we wanted to have the operations executed in a different order; we can use parentheses to group everything, like so:

```python
print(((2*3)**5) + 1)
```

Expressions within parentheses take precedent over everything else, so when in doubt group with ()!

## Variables

If instead of displaying results you wish to store them for later use, you'll need to use the assignment operator: '='. The '=' symbol takes the value or output of the expression on the right and stores it in the variable on the left. Unlike the operators we've already seen, don't confuse equality in the mathematical sense with what the assignment operator '=' does. For example:

```python
u = 1 + 2  # u is a 'variable' that contains the result of the operation 1 + 2, which is 3.
print(u)   # When 'print' is used on a variable, the variable contents is what gets displayed.
print(2*u) # Variables can also used in mathematical expressions; again the contents of the variable is what gets used.
v = u**2   # You can even use expressions with variables to assign values to other variables. 
print(v)   # v contains the square of the contents of u. Crucially, this assignment DOES NOT change the contents of u.
u = u + 1  # This, however, will change the contents of u. First, the pre-existing contents of u will have a 1 added to
           # them. Then, u's contents will be changed to the whatever the result of that operation was. In this case,
           # we've incremented u by 1.
print(u)
```

## Plotting with `MatPlotLib`

Simple plots in Python can be done using the matplotlib package, with help from the numpy package:

```python
import matplotlib.pyplot as plt # The 'import' keyword is used to include 'packages' that are not part of 
import numpy as np              # default python library. The 'as' keyword allows us to rename the imported package 
                                # for use in our program. If we didn't rename the import, we would need to reference
                                # it with 'matplotlib.pyplot'. We also bring in the numpy package, which is optimized
                                # for many mathematical operations.
        
x = np.linspace(-np.pi,np.pi,1000) # This creates a 'vector' of 1000 equally-spaced points between -pi and pi.
                                   # More on vectors later.
    
plt.plot(x,np.sin(x), 'k-') # This statement actually plots what we want. The first argument to the function is a set
                            # of x values, the second is a set y values (in this case we are taking the sine of every
                            # point in our set of x's), and the last argument is a what's known as a 'format string',
                            # it tells the function how to draw line itself. The 'k' markers black, and the '-' makes
                            # connects the markers with lines.
```

The plot we have above isn't bad, but we can do better by using other 'matplotlib' commands:

```python
plt.figure(figsize=(16,12)) # the 'figure' function allows you to set specific attributes of the figure. Here we only
                            # use the figsize attribute to make the plot bigger. This plot is now 16 inches wide and
                            # 12 inches tall.
        
plt.plot(x,np.sin(x), 'k-', label="$\sin(x)$", linewidth=5) # 'label' allows you to name each curve on the plot,
plt.plot(x,np.cos(x), 'b-', label="$\cos(x)$", linewidth=5) # this may be necessary if you intend to plot more than 
                                                            # one curve on your plot, like below. 'linewidth' does
                                                            # what you might expect.
        
plt.xlabel("$x$", size=30) # 'xlabel' and 'ylabel' allow you to label the x and y axis, respectively. 
plt.ylabel("$y$", size=30) # 'size' allows you to choose the fontsize for the label. 
                           # *fontsize changed*


plt.xticks(size=30) # *set the fontsize of xticks *
plt.yticks(size=30) # *set the fontsize of yticks *

plt.title("$\sin(x)$ and $\cos(x)$ plotted on $x \in [-\pi,\ \pi]$", size =30) # 'title' allows you to set the title
                                                                               # of the whole plot.
                                                                               # *fontsize changed*
    
plt.legend(loc=('upper left'), framealpha=0,fontsize = 30) # If any of your curves are labeled, 'legend' allows you to
                                                          # have an inset legendon your plot.
                                                          # *set box off and fontsize*
        
ax=plt.gca() # *Get the handle of the axes*
ax.spines['bottom'].set_linewidth(2) # *set linewidth of axes*
ax.spines['left'].set_linewidth(2) # *set linewidth of axes*
ax.spines['right'].set_linewidth(2) # *set linewidth of axes*
ax.spines['top'].set_linewidth(2) # *set linewidth of axes*

ax.tick_params(direction='in', width=2, length=8, right=True, top=True) # *set properties of axes ticks*
```

```python
from mpl_toolkits import mplot3d # We import the mplot3d from mpl_toolkits, another python package necessary for 
                                 # plotting in 3D, such an contour plots.

def f(x, y):
    return np.sin(-x)*np.exp(-y**2) # This define a function that accepts 2 arguments and returns a single one.
                                    # Much like the np.sin and np.cos functions, above we can evaluate this function
                                    # over whole sets of x's and y's. More on that when we get to vectors.

x = np.linspace(-np.pi, np.pi, 100) # We'll use these vectors for the 3d plot below.  
y = np.linspace(-1, 1, 100)

X, Y = np.meshgrid(x, y) # np.meshgrid creates a structure that associates each point in an x-y plane with
                         # with a set of components that correspond to the coordintes of x and y at that point.

Z = f(X, Y) # Z contains f evaluated at ever point in the meshgrid X, Y.


# The code below uses the X, Y, and Z variables from above to create the contour plot below.
# Play with the code and see what you can change based on what we've already seen!
plt.figure(figsize=(16,12)) 
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 200, cmap='binary')
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z");
```

## Symbolic calculus using SymPy

### Differentiation

```python
import sympy # The sympy package is used for symbolic operations, like finding derivatives and indefinite integrals.

x, y, z = sympy.symbols('x y z') # sympy.symbols creates a set of 'symbolic' variables, analogous to variables in
                                 # the normal calculus sense. 
```

```python
sympy.diff(sympy.ln(x), x) # sympy.diff allows you to compute derivatives. The first argument is the expression
                           # whose derivative you want to find and the second argument is the variable you are 
                           # taking the derivative with respect to.
```

```python
sympy.diff(sympy.exp(sympy.cos(y)**2), y)
```

```python
r = sympy.sqrt(x**2 + y**2 + z**2) # Once you have symbolic variables, you can create complex expression and 
                                   # assign them to other variables.
    
sympy.diff(r, x, y, z) # If you're expression involves multiple variables you can also compute partial derivatives,
                       # even with respect to multiple variables at once. This is equivalent to taking the partials
                       # of r with respect to x, y, and z.
```

```python
fPrime = sympy.Derivative(r, x, y, z) # sympy.Derivative gives the symbolic expression before the derivative 
                                      # is evaluated. Assigning it to a variable allows you to evaluate it later.
fPrime
```

```python
fPrime.doit() # doit allows you to evaluate fPrime. 
```

```python
f = (x*y)**z
f.diff((z, 2)) # Calling diff on a saved expression allows you to take higher order derivatives. For example,
               # this takes the second order derivative of f with respect to z.
```

```python
expx = sympy.exp(-x)
iexpx = sympy.integrate(expx, x)  # sympy.integrate allows you to compute the indefinite integral of the expression;
                                  # however, the constant of integration is left off. 
```

```python
sympy.Integral(expx, x)# Similar to sympy.Derivative, sympy.Integral gives you the integral before evaluation.
```

```python
sympy.integrate(expx,(x, 0, np.inf)) # If the expression has an elementary antiderivative, then including bounds
                                     # will compute the definite integral.
```

```python
sympy.integrate(sympy.exp(sympy.cos(x)), x) # If the expression does not have an elementary antiderivative, then
                                            # you will get back an unevaluated integral. There are ways that can 
                                            # Python can numerically integrate expressions like this one, but that
                                            # is beyond the scope of this chapter.
```

## Vectors, Matrices, and Linear Algebra
The following sections makes heavy use of the numpy package, due to it's powerful central construct: the numpy n-dimensional array or ndarray.

*A small note on vocabulary*: Generally, the word vector is associated with, and analogous to, a 1-dimensional array. In spite of this, most of the things we have to say about vectors also apply to higher dimensional arrays and by that same token it may be the case that the terms 'matrix' and '2-dimensional array' are also interchangeable.
### The ndarray

```python
v = np.array([1,2,3]) # np.array creates a ndarray with the given list as elements. For example, 
                      # this creates a row vector that contains the elements [1, 2, 3].
    
print("v:",v)       # The print function can be used on vectors to display their contents.
print("shape of v:",v.shape) # All numpy arrays have a 'shape' attribute that tells you how many dimensions
               # the array has and how large those dimensions are. In this case, v is a 
               # 1-dimensional array of length 3.
        
print("transpose of v:",np.transpose(v)) # np.transpose gives the tranpose of the array. In this case, the transpose
                                         # does nothing because v is a 1-dimensional array.
    
v = np.array([v])                # However, redefining v in this way turns it into a 2-dimensional array, where 1
print("new v:",v)                # dimension is of length 1 and the other is of length 3; which is reflected in the
print("shape of new v:",v.shape) # new output of v.shape.

print("transpose of new v:\n",np.transpose(v)) # Since v is now a 2-dimensional array, we also get the transpose 
                                               # we expect.

m = np.array([[1,2],[3,4]]) # You can also create 2-dimensional arrays that are analogous to matrices. 
print("m:\n",m)
print("shape of m:",m.shape)
print("transpose of m:\n",np.transpose(m))
```

### Accessing array elements and array slices
After defining a vector or matrix, it may be necessary to access specific elements. This is done with the indexing operator [ ].

```python
d = np.random.randint(-5, 5, size=10) # This is a method to generate 10 random integers between -5 and 5.
print("d:",d)                         # More on the 'random' module later.

print("first element of d:",d[0])        # Using the indexing operator on an array allows you to access specific 
print("last element of d:", d[d.size-1]) # elements of the array, each element has it's own index in the array. 
                                         # Indexing starts at 0 and ends at the length of the vector minus 1. 
                                         # Trying to access indices outside of this range for any given array
                                         # will cause an error.
            
print("last element of d:", d[-1]) # You can also get the last element of the vector by indexing '-1'.
```

You can also access subsets of a vector using array slicing with the indexing opertor.

```python
print("first three elements of d:", d[0:3]) # The ':' symbol allows you to specify a range of indices to
print("first three elements of d:", d[ :3]) # access from the array. Leaving off the first index will access
print("last three elements of d:", d[7: ])  # all elements up to the second index, and leaving off the second
print("d:", d[:])                           # accesses all elements including and after the first. Leaving off 
print("every third element of d:", d[::3])  # both access the whole array. You can also give a third number to
                                            # specify an increment.
    
print("First row of m:",m[0,:])     # Array indexing and slicing can also be done on matrices, but instead you index
print("Second column of m:",m[:,1]) # in multiple dimension, like so: m[d1, d2].
```

### Vectorization
One of the powerful tools that the ndarray provides is something called vectorization, which allows you to perform mathematical operations on entire vectors as opposed to doing those operations on each element of the vector sequentially.

```python
u = np.arange(0,10)  # np.arange creates a sequence of integers from the first argument to the last, with the last
w = np.arange(10,20) # not being included; so u is [0, ..., 9] and w is [10, ... , 19], they are also 1-dimensional 
l = u + w            # numpy ndarrays. (just about all arrays that you get from numpy functions are ndarrays)

print("u + w:",l) # This is a vectorized sum; in other words, every element of u and w that are in the 
                      # same position are added together and a new vector is created from these pairwise additions. 
                      # Addition is not special in this respect, though, as subtraction, multiplication, division,
                      # and even exponentiation support vectorization.
print("u - w:",u - w)
print("u*w:",u * w)
print("u/w:",u / w)
print("u**w:",u**w)
    
m1 = np.arange(0,100).reshape(10,10)   # All numpy arrays also have the reshape function, which allows you to turn
m2 = np.arange(100,200).reshape(10,10) # a vector into a 2-dimensional array or matrix. Note that the length of the
                                       # vector must equal the product of the dimensions of the matrix for reshape 
                                       # to work properly.

print("m1 + m2:\n", m1 + m2) # In spite of it being called 'vectorization' it also applies to matrices as well. 
                             # The key thing to bear in mind with vectorization between ndarrays is that the 2
                             # ndarrays involved must be of the length for vectors or of the same dimensions for 
                             # higher dimensional arrays. 
```

Vectorization is not just restricted to operations between vectors or matrices though, it can used with scalars as well:

```python
print("u:",u)         # Here we have the u vector from the previous cell, and then the vectorized sum between 2 and u.
print("2 + u:",2 + u) # Unlike sums between vectors, instead 2 is added to each element of u and like the previous
                      # case there is nothing special about addition here.
    
print("2 - u:",2 - u)
print("2*u:",2*u)
print("u/2:",u/2)
print("2**u:",2**u)
```

Other numpy functions also support vectorization, as we saw during the plotting portion:

```python
print("e**u:\n",np.exp(u))      # The exponential function.
print("sin(u):\n",np.sin(u))    # Sine of u.
print("cos(u):\n",np.cos(u))    # Cosine of u. Note that for trig functions, the arguments are radians. 
print("ln(u+1):\n",np.log(u+1)) # The natural logarithm. Note that log2 and log10 also exist.
                                # Further, each of these functions support scalar arguments, as well.
```

### Matrix Multiplication

```python
print("m1@m2:\n",m1@m2) # The '@' symbol is used for conventional matrix multiplication. As such it follows the 
print("m1@u:",m1@u)     # definition of matrix multiplication that we find in Linear Algebra regarding the 
                        # dimensions of the 2 operands.

print("shape of m1:",m1.shape) # If you're unsure whether a multiplication is possible, check the dimensions
print("shape of m2:",m2.shape) # using the shape attribute.
print("shape of u:",u.shape)
```

### Linear Algebra
In addition to what we've already seen, numpy also has a robust set of linear algebra functions.

```python
m = np.random.randint(-10, 10, size=(3,3)) # Here we use the random module again to create a matrix and 
b = np.random.randint(-10, 10, size=3)     # vector to use the linear algebra functions on.

print("\nm:\n",m)
print("\nb:",b)
print("\nDeterminant of m:",np.linalg.det(m)) # np.linalg.det computes the determinant of the matrix.

print("\nx such that mx=b:",np.linalg.solve(m,b)) # np.linalg.solve solves the linear system of [m |b].

print("\nDiagonal matrix of b:\n", np.diag(b)) # np.diag creates a diagonal matrix that has the elements of b on
                                               # it's diagonal. Note that this function only works on vectors.

print("\nInverse of m:\n",np.linalg.inv(m)) # np.linalg.inv computes the inverse of the matrix. Note that for
                                            # singular matrices, this function will have an error. 
    
print("\nm@mInverse:\n",m@np.linalg.inv(m)) # Here we verify that the inverse is the in fact the inverse. The
                                            # result may not exactly be the identity due to round off errors.

print("\nIdentity Matrix of dimension 3:\n", np.identity(3)) # np.identity creates an identity matrix of the
                                                             # same dimension as the integer argument.

e,ev = np.linalg.eig(m)            # np.linalg.eig computes the eigenvalues and eigenvectors of a matrix.
print("\nEigenvalues of m:\n",e)   # Each are returned as separate arrays, so we store each in a variable 
print("\nEigenvectors of m:\n",ev) # to display them individually.

U, S, V = np.linalg.svd(m)                 # np.linalg.svd computes the singular value decomposition of a matrix.
print("\nLeft-singular matrix of m:\n",U)  # Like the previous function, this one returns multiple arrays, so we 
print("\nSingular values of m:\n",S)       # store each in it's own variable and display them in turn.
print("\nRight-singular matrix of m:\n",V)

print("\nDot product of b and (b + 1):", np.dot(b,b+1)) # np.dot computes the dot product of 2 vectors; alternatively,
                                                        # the dot product can be computed using normal matrix 
                                                        # multiplication and np.transpose.
```

## Statistics with NumPy
For statistical purposes, we have some functions built into numpy and we'll use them along with MatPlotLib.

```python
x = np.random.normal(loc=0, scale=1,size=10) # The np.random module allows you to generate random values within a 
                                             # given probability distribution. In this case, we generate 10 random 
                                             # values within a normal distribution mean 0 and standard deviation 1. 
                                             # Some function support keyword arguments, which is what we have here.
                                             # Instead of passing arguments by position, you can specify by name.

print("x:",x)

print("\nmean(x):",np.mean(x)) # np.mean computes the arithmetic mean of the vector x. 

print("\nvar(x):",np.var(x)) # np.var computes the variance of the vector x.

print("\nstd(x):",np.std(x)) # np.std computes the standard deviation of the vector x.

print("\nmedian(x):",np.median(x)) # np.median computes the median of the vector x.

print("\nquantiles(x):",np.quantile(x,[0,.25,.50,.75,1])) # np.quantile computes percentiles of a vector x. In 
                                                          # addition to the vector itself, it is necessary to
                                                          # provide the percentiles either individually, or as 
                                                          # a list. This example return a vector of the quantiles.

print("\nmin(x):",np.min(x)) # np.min returns the least value of the vector x. 

print("\nmax(x):",np.max(x)) # np.max returns the greatest value of the vector x.
```

The cells below show how we can use MatPlotLib for statistical purposes along with normal mathematical plotting.
