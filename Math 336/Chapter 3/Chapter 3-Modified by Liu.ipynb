### DataFrames, reading file input, and pandas!
Often, it is necessary to read information from a file to analyze it. For that, one of the more useful python packages is 'pandas'. pandas has functions that allow for the input of files from your computer, but also allows you to store the data in something called a DataFrame, which is a data structure that is similar to a 2d array but with other 'Quality of Life' features that make grouping and indexing the data easier.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# In addition to the import of pandas as pd, we've also imported numpy and matplotlib. Both of those packages
# have immensensely useful tools, as stated in chapter 2, so you'll very likely see them in every chapter from
# here on out.

globalLifeExpectancy = pd.read_csv("API_SP.DYN.LE00.IN_DS2_en_csv_v2_1120968.csv", header=2, index_col=0)
globalLifeExpectancy
# Above we use our pandas import to read a csv file (csv is short for Comma Separated Values, a very common
# file type for data storage) and store it's data in a DataFrame called globalLifeExpectancy. The first argument
# to the function is the name of the file (given as a string), the second argument is a keyword argument specifying
# the number of lines in the 'header' of the file. Some csv files will have multiline headers that explains the data
# in the file, and in this case there are 2. The third argument specifies which column's data will be used to label
# each row, more on these 'labels' momentarily.

# The csv file in question can be found at: https://data.worldbank.org/indicator/SP.DYN.LE00.IN
# Just scroll down, click CSV under Download and extract the file to the same folder as this 
# Jupyter Notebook file.
```

Below, we extract the life expectancy data for France and find a best fit line to the data.

```python
franceData = np.array(globalLifeExpectancy.loc['France', '1960':'1989'])
years = np.arange(1960,1990)
# Above we extract the data associated France from 1960 to 1989 and store it in a numpy array called franceData.
# Along with that, we create a range of integers from 1960 to 1989 and store it in a numpy array called
# franceData.

fTrend = np.polyfit(years, franceData.astype(np.float), 1)
fTrend[0], fTrend[1]
# The np.polyfit function generates the coefficients of a polynomial that is a best fit for a set of data.
# In this case, we provide the x values (years), y values (franceData), and the degree of polynomial we want (1),
# which is a linear best fit.
```

```python
def linear_model(x):
    return lambda t: x[0]*t + x[1]

# This function accepts an array of coefficients and returns a linear 
# lambda function with those coefficients.
```

### Linear models and line of best fit.

Now we want to plot the data we extracted from the DataFrame, along with our line of best fit. We also include some plt functionality to improve the look of the plot.

```python
franceLine = linear_model(fTrend)
plt.figure(figsize=(12,9))
plt.scatter(years, franceData, s=70, c='k', alpha=1, marker='o', zorder=10) # *use scatter to change the symbol properties*
plt.plot(years, franceLine(years), 'b-',label=f"Upward trend of {fTrend[0]:.3f} per year", linewidth=2,zorder=5) # *set the order of plot*
plt.xlabel("Year", size=20)
plt.ylabel("Life Expectancy", size=20)
plt.tick_params(length=6, width=2, labelsize=20) # This allows you to modify the size and number of 'ticks'
                                                 # along the axis
plt.legend(loc='upper left', fontsize=15)
plt.title("Life Expectancy for France 1960-1989 with linear trend.", fontsize=20); # *fontsize changed*

ax=plt.gca() # *Get the handle of the axes*
ax.spines['bottom'].set_linewidth(1.3) # *set linewidth of axes*
ax.spines['left'].set_linewidth(1.3) # *set linewidth of axes*
ax.spines['right'].set_linewidth(1.3) # *set linewidth of axes*
ax.spines['top'].set_linewidth(1.3) # *set linewidth of axes*
```

Now we look at a linear model between the heating degree day (HDD) and energy consumption (kWh) by month. Background information on these metrics is available in the textbook on page 36.

```python
HDD = np.array([163, 228, 343, 373, 301, 238, 137, 84, 38, 15, 14, 34])
kWh = np.array([593, 676, 1335, 1149, 1127, 892, 538, 289, 172, 131, 134, 134])

HDDbykWh = pd.DataFrame(data=[HDD,kWh], index=['HDD','kWh'], 
                columns=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
HDDbykWh
```

The creation of the DataFrame above was completely unecessary for what we are going to do next and is only there to show you that you can create your own from data entered manually or data that is already stored in an array. As already mentioned, DataFrames have a lot of quality of life features that you are encouraged to look into yourself. For now, lets find the linear trend and correlation between HDD and kWh:

```python
cTrend = np.polyfit(HDD, kWh, 1)
R = np.corrcoef(HDD, kWh)[0,1]**2
plotlabel1 = (
    f"Linear Model: kWh = {cTrend[0]:1.3f}HDD + {cTrend[1]:2.3f}\n"
    f"Correlation Coefficient: R = {R:.4f}"
    )
```

```python
hddLine = linear_model(cTrend)
plt.figure(figsize=(12,9))
plt.scatter(HDD, kWh, s=80, c='k', alpha=1, marker='o', zorder=10) # *use scatter to change the symbol properties*
plt.plot(HDD, hddLine(HDD), 'r-', label=plotlabel1)
plt.xlabel("Climate Index Data: HDD", size=20)
plt.ylabel("Energy Index Data: kWh", size=20)
plt.tick_params(length=6, width=2, labelsize=20) 
plt.legend(loc='upper left', fontsize=15)
plt.title("Monthy household energy consumption vs. Climate data", fontsize=20); # *fontsize changed*

ax=plt.gca() # *Get the handle of the axes*
ax.spines['bottom'].set_linewidth(1.3) # *set linewidth of axes*
ax.spines['left'].set_linewidth(1.3) # *set linewidth of axes*
ax.spines['right'].set_linewidth(1.3) # *set linewidth of axes*
ax.spines['top'].set_linewidth(1.3) # *set linewidth of axes*
```

Now we move onto another data set on climate change.

```python
dtmean = pd.read_csv("HansenGlobalT.txt", sep="\s+", header=0, index_col=0)
# We are still using the read_csv function from pandas, but you might notice that the file we're reading
# in is not a CSV. Fear not! Pandas functions are versatile, and we can use the same function for this file
# as long as we specify the separator we're using to read in the data; this separator in question, sep="\s+",
# is what's known as a 'regular expression' for one or more spaces.
dtmean
```

Now that we have the data, lets extract in for processing. Below we have a function that I wrote to display a Summary of the data of an array.

```python
def Summary(x):
    deets = np.quantile(x, [0, .25, .5, .75, 1])
    deets = np.array([np.append(deets, np.mean(x))])
    df = pd.DataFrame(data=deets, columns=['Min.', '1st Qu.', 'Median', '3rd Qu.', 'Max.', 'Mean'],
                      index=['values'])
    print(df.to_string())
```

The Summary function displays the quantiles and the mean of the input vector.

```python
yrstime = np.arange(1880, 2015)
dtAnom = np.array(dtmean.loc[:,'Anomaly'])
Summary(dtAnom) # Here we find the Summary of dtAnom and yrstime.
Summary(yrstime)
```

```python
hTrend = np.polyfit(yrstime, dtAnom.astype(np.float), 1)
hTrend[0], hTrend[1]
```

```python
anomLine = linear_model(hTrend)
plt.figure(figsize=(12,9))
plt.plot(yrstime, dtAnom, 'ko-')
plt.plot(yrstime, anomLine(yrstime), 'r-', label=f"Linear Trend: {hTrend[0]*100:.2f} $\degree$C per century", linewidth=4)
plt.xlabel("Year", size=20)
plt.ylabel("Temperature Anomalies [$\degree$C]", size=20)
plt.tick_params(length=6, width=2, labelsize=20) 
plt.legend(loc='upper left', fontsize=15)
plt.title("Global Annual Mean Land Surface Air Temperatue Anomalies", fontsize=20);
```

We've already seen how we can get an idea of the overall trend of the data with a line of best fit,
but what if we want to determine the linear trend for subsets of the data and see how the trend itself changes?
That's what we'll get into next.

```python
plt.figure(figsize=(12,9))
plt.plot(yrstime, dtAnom, 'ko-')

fmt = ['b-', 'g-', 'c-', 'm-', 'r-']
indices = [31, 71, 96, 121, 150]
endyear =[1910, 1950, 1975, 2000, 2014]
# These lists contain bits of data that we'll loop through in the for loop below.
# Unlike the already touched on numpy arrays, lists are a data structure built
# directly into python. They're not as versatile or as optimized as numpy arrays,
# but for a small number of data that aren't being operated on directly they suffice.

for (col, i, year) in zip(fmt, indices, endyear):
    mTrend = np.polyfit(yrstime[:i], dtAnom.astype(np.float)[:i], 1)
    temporalLine = linear_model(mTrend)
    lab = f"1880-{year} Trend: {mTrend[0]*100:.2f} $\degree$C per century"
    plt.plot(yrstime[:i], temporalLine(yrstime[:i]), col, linewidth=4, label=lab)
# The structure above is called a for loop, and they're used when it's necessary to do a particular
# operation multiple times with (pontentially, but not necessarily) different input on each loop around.

plt.xlabel("Year", size=20)
plt.ylabel("Temperature Anomalies [$\degree$C]", size=20)
plt.tick_params(length=6, width=2, labelsize=20) 
plt.legend(loc='upper left', fontsize=15)
plt.title("Global Annual Mean Global Surface Air Temp from Hansen", fontsize=15);
```
