### `matplotlib` plot parameters and style dictionaries.

the look of your plots and graphs. In order to maintain a certain level of
consistency between plots chapter to chapter, we'll use something called a
*style dictionary*. Basically, it's a way to choose a set of default
parameters for all plot in a given `python` script or, in our case, a
`python` notebook. Using this allows you to make plot quite pretty by
default, and any additional changes that need to be made on a per-plot basis
can still be implemented.

```python
styledict = {'xtick.labelsize':25,
             'xtick.major.size':9,
             'xtick.major.width':1,
             'ytick.labelsize':25,
             'ytick.major.size':9,
             'ytick.major.width':1,
             'legend.framealpha':0.0,
             'legend.fontsize':15,
             'axes.labelsize':20,
             'axes.titlesize':25,
             'axes.linewidth':2,
             'figure.figsize':(12,8),
             'savefig.format':'jpg'}
# Our style dictionary is defined above. See if you can deduce which keywords
# control which plot property.

plt.rcParams.update(**styledict)
# In order for our new changes to be applied, we have to update matplotlib's
# 'rcParams'; the list of properties that plt accesses by default. The line of
# code above does just that.
```

Below, we extract the life expectancy data for France and find a best fit
line to the data.

```python
franceData = np.array(globalLifeExpectancy.loc['France', '1960':'1989'])
years = np.arange(1960,1990)
# Above we extract the data associated France from 1960 to 1989 and store
# it in a numpy array called franceData. Along with that, we create a range
# of integers from 1960 to 1989 and store it in a numpy array called
# years.

fTrend = np.polyfit(years, franceData.astype(np.float), 1)
# The np.polyfit function generates the coefficients of a polynomial that is
# a best fit for a set of data. In this case, we provide the x values (years),
# y values (franceData), and the degree of polynomial we want (1), which is a
# linear best fit.

print(f"Slope of linear trend: {fTrend[0]:5.4f}, "
      f"y-intercept of linear trend: {fTrend[1]:5.2f}")
```

```python
def linear_model(x):
    """
    This function accepts a numpy array of length 2, and returns a linear
    lambda function using the elements of the array as the coefficients,
    like so: f(t) = x[0]*t + x[1]
    """
    return lambda t: x[0]*t + x[1]

# This function accepts an array of coefficients and returns a linear 
# lambda function with those coefficients.
```

### Linear models and line of best fit.

Now we want to plot the data we extracted from the DataFrame, along with our
line of best fit. We also include some plt functionality to improve the look
of the plot.

```python
franceLine = linear_model(fTrend)
plt.figure(figsize=(12,9))
plt.plot(years, franceData, 'ko')
plt.plot(years, franceLine(years), 'b-',
         label=f"Upward trend of {fTrend[0]:.3f} per year")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.tick_params(direction='in')
# This allows you to modify the direction in which the 'ticks' protrude along
# the axis.

plt.legend(loc='upper left')
plt.title("Life Expectancy for France 1960-1989 with linear trend.");

plt.savefig("CH3;LifeExpectancyFranceTrend.jpg")
```

Now we look at a linear model between the heating degree day (HDD) and energy
consumption (kWh) by month. Background information on these metrics is available
in the textbook on page 36.

```python
HDD = np.array([163, 228, 343, 373, 301, 238, 137, 84, 38, 15, 14, 34])
kWh = np.array([593, 676, 1335, 1149, 1127, 892, 538, 289, 172, 131, 134, 134])

HDDbykWh = pd.DataFrame(
    data=[HDD,kWh], index=['HDD','kWh'],
    columns=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
             'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
# The DataFrame created above is made from data we defined instead of data from a
# file (like the one at the top of this notebook); as such, `index` and `columns`
# allows us to label the rows and columns of the DataFrame instead of extrapolating
# them from a file.

HDDbykWh
```

The creation of the DataFrame above was completely unnecessary for what we are
going to do next and is only there to show you that you can create your own
from data entered manually or data that is already stored in an array. As
already mentioned, `DataFrame` has a lot of quality of life features that
you are encouraged to look into yourself. The `DataFrame` API can be found
[here](https://pandas.pydata.org/docs/reference/frame.html). For now, lets
find the linear trend and correlation between HDD and kWh:

```python
cTrend = np.polyfit(HDD, kWh, 1)
# cTrend is our set of linear fit coefficients

R = np.corrcoef(HDD, kWh)[0,1]**2
# R is the coefficient of correlation between HDD and kWh.

plotlabel1 = (
    f"Linear Model: kWh = {cTrend[0]:1.3f}HDD + {cTrend[1]:2.3f}\n"
    f"Correlation Coefficient: R = {R:.4f}"
    )
# plotlabel1 will be used as a label for the plot below.

hddLine = linear_model(cTrend)
# hddLine is a linear function that outputs the linear trend between HDD and kWh
```

```python
plt.figure(figsize=(12,9))
plt.plot(HDD, kWh, 'ko')
plt.plot(HDD, hddLine(HDD), 'r-', label=plotlabel1)
plt.xlabel("Climate Index Data: HDD")
plt.ylabel("Energy Index Data: kWh")
plt.legend(loc='upper left')
plt.title("Monthly household energy consumption vs. Climate data");

plt.savefig("CH3;MonthlyEnergyConsumptionTrend.jpg")
```

Now that we have the data, lets extract it for processing. Below we have a
function that I wrote to display a Summary of the data of an array.

```python
def Summary(x):
    """
    This function accepts a 1-D numpy array as an argument, and uses a
    DataFrame to return a string that displays a summary of the data:
    the quantiles and the mean.
    """
    deets = np.quantile(x, [0, .25, .5, .75, 1])
    deets = np.array([np.append(deets, np.mean(x))])
    df = pd.DataFrame(data=deets,
                      columns=['Min.', '1st Qu.', 'Median',
                               '3rd Qu.', 'Max.', 'Mean'],
                      index=['values'])
    print(df.to_string())
```

The Summary function displays the quantiles and the mean of the input vector.

```python
yrstime = np.arange(1880, 2015)
dtAnom = np.array(dtmean.loc[:,'Anomaly'])
print("Summary of dtAnom")
Summary(dtAnom)
print("Summary of yrstime")
Summary(yrstime)
```

We've already seen how we can get an idea of the overall trend of the data with a line of best fit,
but what if we want to determine the linear trend for subsets of the data and see how the trend itself changes?
That's what we'll get into next.

```python
fmt = ['b-', 'g-', 'c-', 'm-', 'r-']
indices = [31, 71, 96, 121, 150]
endyear =[1910, 1950, 1975, 2000, 2014]
# These lists contain bits of data that we'll loop through in the for loop below.
# Unlike the already touched on numpy arrays.
```

```python
plt.figure(figsize=(12,9))
plt.plot(yrstime, dtAnom, 'ko-')

for (col, i, year) in zip(fmt, indices, endyear):
    mTrend = np.polyfit(yrstime[:i], dtAnom.astype(np.float)[:i], 1)
    temporalLine = linear_model(mTrend)
    lab = f"1880-{year} Trend: {mTrend[0]*100:.2f} $\degree$C per century"
    plt.plot(yrstime[:i], temporalLine(yrstime[:i]), col,
             linewidth=4, label=lab)
# zip() allows you to take an arbitrary number of iterables and turn them
# into one list of tuples element by element; which allows you to have
# multiple for loop indices.

plt.xlabel("Year")
plt.ylabel("Temperature Anomalies [$\degree$C]")
plt.legend(loc='upper left')
plt.title("Global Annual Mean Global Surface\nAir Temp from Hansen");

plt.savefig("CH3;VariableLandAirTempAnomaliesTrend.jpg", bbox_inches='tight')
```
