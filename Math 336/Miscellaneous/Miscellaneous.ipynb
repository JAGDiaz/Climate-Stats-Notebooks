### Phase Diagrams
Suppose you wanted to characterize the behavior of the following
second-order linear homogeneous differential equation:
$$y''(t) +y'(t) - 6y(t) = 0$$
The solution to this equation is $y(t) = c_1e^{-3t} + c_2e^{2t},\
c_1=y(t_0),\ c_2=y'(t_0)$, which can be found using a variety of
methods. This second-order linear differential equation can also be
phrased as a system of first-order linear differential equations by
defining a new set of variables, which we'll denote $y_1,\ y_2$:
\begin{align*}
y_1 = y &\implies y_1' = y' = y_2 \\
y_2 = y_1' &\implies y_2' = y'' = 6y_1 - y_2
\end{align*}
So we now have the system:
\begin{array}{l}
y_1' = y_2 \\
y_2' = 6y_1 - y_2
\end{array}
Given these definitions, we can re-express this system as matrix-vector
multiplication; like so:
$\dot{\vec{y}} = A\vec{y}$ with $A = \left[\begin{matrix}
0 & 1 \\
6 & -1
\end{matrix}\right]$

We already have a analytical solution to this equation, but there is more
to examine than just that here. To determine the dynamics of the system,
we'll need to create a phase diagram.

```python
a = 6
arrs = 25
y1, y2 = np.meshgrid(np.linspace(-a, a, arrs), np.linspace(-a, a, arrs))
y1dot = y2
y2dot = 6*y1 - y2
```

```python
plt.quiver(y1dot, y2dot)
```
